# Pandas Indexing & Slicing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/pandas/pokemon_data.csv')
df
```

```python
# iloc() ==> index_location
# for reading ROWS
df.iloc[2:5]
```

```python
# Rows of a column
for index, row in df.iterrows():
    print(index, row['Name'])
```

## Selective Search

Is a concept where you search rows at specific location for given condition

```python
# loc() ==> location
# dataframe.loc[dataframe['Column Name'] == 'Value to search']
df.loc[df['Type 1'] == 'Grass']
```

```python
# other ftn can be performed as well
df.loc[df['Type 1'] == 'Grass'].tail(20)
```

# Features & Target

Target ==> Is the attribute whose predication is to be done
Features ==> Are the attributes which are used to make predication of given target

Both are columns name.

80-90% of the times you are given a .csv file to analyze but here I'm using a .xlsx file

```python
# Mostly .csv files are given for this purposes
hiring = pd.read_excel('datasets/hiring.xlsx')
hiring
```

Now, we will predict the salary of a person based on given attributes like exp, test score etc.

So, here 'Salary' is our 'Target' and other attributes are our 'Features'
### Note
Target is taken on x-axis

Features are taken on y-axis

```python
x = hiring['salary']
y = hiring[['experience','test_score','interview_score']]
```
